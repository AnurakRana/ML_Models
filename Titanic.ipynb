```python
df=pd.read_csv('/kaggle/input/titanic/train.csv')
```

```python
df.head()
```

```python
df['Cabin'].fillna('U', inplace=True)
df['Cabin'] = df['Cabin'].apply(lambda x: x[0])
```

```python
df['Cabin'].head()
```

```python
df['Survived'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
```

```python
df['Sex']=encoder.fit_transform(df['Sex'])
df['Embarked']=encoder.fit_transform(df['Embarked'])
df['Cabin']=encoder.fit_transform(df['Cabin'])
```

```python
df.describe()
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
df['Age'].fillna(df['Age'].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df=df.select_dtypes(include=['number'])
```

```python
df.corr()
```

```python
corr_matrix = df.corr()
selected_features = corr_matrix["Survived"].abs().sort_values(ascending=False)
selected_features = selected_features[selected_features >= .1].index.tolist()
selected_features.remove("Survived")
print(selected_features)
```

```python
from imblearn.over_sampling import SMOTE
smote=SMOTE()
X=df[selected_features]
y=df['Survived']
X_smote, y_smote = smote.fit_resample(X, y)
df_smote = pd.concat([pd.DataFrame(X_smote, columns=X.columns), pd.DataFrame(y_smote, columns=['Survived'])], axis=1)
print(df_smote['Survived'].value_counts()) 
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_smote,y_smote,test_size=0.2,random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
pred=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy=accuracy_score(pred,y_test)
```

```python
accuracy
```
